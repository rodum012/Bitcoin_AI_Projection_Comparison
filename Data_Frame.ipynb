```python
BTC_df = get_data("BTC-USD", start_date="6/30/2019", end_date="06/30/2022", index_as_date = True, interval="1d")
Dodge_df = get_data("DOGE-USD", start_date="6/30/2019", end_date="06/30/2022", index_as_date = True, interval="1d")
Eth_df = get_data("ETH-USD", start_date="6/30/2019", end_date="06/30/2022", index_as_date = True, interval="1d")
Ada_df = get_data("ADA-USD", start_date="6/30/2019", end_date="06/30/2022", index_as_date = True, interval="1d")
Xrp_df = get_data("XRP-USD", start_date="6/30/2019", end_date="06/30/2022", index_as_date = True, interval="1d")

Top5 = [BTC_df,Dodge_df,Eth_df,Ada_df,Xrp_df]
Total_df = pd.concat(Top5, axis=1)

Total_df.head()
```

```python
btc_df = Total_df.iloc[:,0:6]
btc_df = btc_df.iloc[:,0:4]
btc_df.head()
```

```python
btc_df = btc_df.iloc[::-1]
btc_df['shifted_close'] = btc_df['close'].shift(-1)
btc_df.dropna(inplace=True)
btc_df.head()
```

```python
btc_df.tail()
```

```python
X = btc_df.iloc[:,0:4]
Y = btc_df['shifted_close']
X.head()
```

```python
X_train, X_test, Y_train, Y_test = X.iloc[:767], X.iloc[767:], Y.iloc[:767], Y.iloc[767:] 
X_train.dropna(inplace=True)
X_test.dropna(inplace=True)
Y_train.dropna(inplace=True)
Y_test.dropna(inplace=True)
```
